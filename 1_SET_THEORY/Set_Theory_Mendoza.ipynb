# 1.Proving Identities

Given: A and B are sets, $\exists$ x such that $(x \in A \wedge x \in B)$  
Show that:

1. $A \setminus B = A \cap B^c$  
$\text{Proof:}$   
$B^c = \mathcal{U} - B \quad \text{(Definition of Complement)}$  
$A \cap B^c = A \cap \mathcal{U} - A \cap B \quad \text{(Distributive Law)}$  
$A \cap \mathcal{U} = A \quad \text{(Domination Law)}$  
$A \cap B^c = A - A \cap B \quad \text{(Domination Law)}$   
$A\setminus B = A \cap B^c = A - A \cap B \quad \text{(Definition of Set Difference)}$

$Q.E.D$

2. $B (B \cap A) \cup (B \cap A^c)$  
$\text{Proof:}$  
$(B \cap A) \cup (B \cap A^c) = B \cap (A \cup A^c) \quad \text{(Distributive Law)}$  
$A \cup A^c = \mathcal{U} \quad \text{(Complement Law)}$  
$B \cap \mathcal{U} = B \quad \text{(Domination Law})$  

$Q.E.D$  

3. $B \setminus A = B \cap A^c$  
$\text{Proof:}$  
$A^c = \mathcal{U} - A \quad \text{(Definition of Complement)}$  
$B \cap A^c = B \cap \mathcal{U} - B \cap A \quad \text{(Distributive Law)}$  
$B \cap \mathcal{U} = B \quad \text{(Domination Law)}$  
$B \cap A^c = B- B \cap A \quad \text{(Domination Law)}$  
$B \setminus A = B \cap A^c = B - B \cap A \quad \text{(Definiition of Set Difference)}$  

$Q.E.D$

4. $A \cap B = A \cup (B \cap A^c)$  
$\text{Proof:}$  
$A \cup (B \cap A^c) = (A \cup B) \cap (A \cup A^c) \quad \text{(Distributive Law)}$  
$(A \cup B) \cap \mathcal{U} \quad \text{(Domination Law)}$

$Q.E.D$

5. $(U^\infty _{i=1} A ^c _{i})^c = \cap^\infty _{i=1} A_{i}$  
$\text{Proof:}$  
$\cap^\infty _{i=1}(A^c _{i})^c \quad \text{(De Morgan's Law)}$  
$\cap^\infty _{i=1} (A^c _{i})^c = \cap^\infty _{i=1} A_i \quad \text{(Double Negation Law)}$  

$Q.E.D$

6. $|A \cup B| = |A| + |B| - |A \cap B|$  
$\text{Proof:}$   
By the definition of the cardinality of a set we count the elements in both set $A$ and set $B$, we then subtract the elements that are counted twice $|A \cap B|$ by the definition of the cardinality of the intersection of sets. Thus, we get $A \cup B = |A| + |B| - |A \cap B|$.

$Q.E.D$

# 3.Set operations

Given:  $U = \{1,2,3,4,5\}$  
$A = \{1,2,3,4,5\}$   
$B = \{4,5,6,7\}$  
$C = \{5,6,7,8,9\}$  
$D = \{1,3,5,7,9\}$  
$E = \{2,4,6,8\}$  
$F = \{1,5,9\}$

$C \oplus D$  

$(C - D) \cup (D - C) = \{6,8\} \cup \{1,3\} = \{1,3,6,8\}$  

$A \setminus B$  

$A - B = \{1,2,3\}$  

$B \setminus A$  

$B - A = \{6,7\}$  

$E \oplus F$

$(E - F) \oplus (D -C) = \{2,4,6,8\} \oplus \{1,5,9\} = \{1,2,4,6,8,9\}$

1. Find the number of people who read at "least one" of the three magazines.
   $\bullet\;|N \cup T \cup F| = |N| + |T| + |F| - |N \cap T| - |N \cap F| - |T \cap F| + |N \cap T \cap F|$  
   $\bullet\;|N \cup T \cup F| = 65 + 45 + 42 - 20 - 25 - 15 + 8$  
   $\bullet\;|N \cup T \cup F| = 100$
     
2. Find the number of pepople who read exactly one magazine.  
   Newsweek $= |N| - |N \cap T| - |N \cap F| + |N \cap T \cap F| = 65 - 20 - 25 + 8 = 28$  
   Time $= |T| - |T \cap F| - |T \cap N| + |T \cap F \cap T| = 45 - 15 - 20 + 8 = 18$  
   Fortune $= |F| - |F \cap N| - |F \cap T| + |F \cap N \cap T| = 42 - 25 - 15 + 8 = 10$
